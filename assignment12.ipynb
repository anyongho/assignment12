```python
import numpy as np
import matplotlib.pyplot as plt
```

## This is the default source code for assignment 12

```python
num     = 1001
std     = 5 

# x  : x-coordinate data
# y1 : (clean) y-coordinate data 
# y2 : (noisy) y-coordinate data

def fun(x):
 
	# f = np.sin(x) * (1 / (1 + np.exp(-x))) 
	f = np.abs(x) * np.sin(x)

	return f

n       = np.random.rand(num)
nn      = n - np.mean(n)
x       = np.linspace(-10,10,num)
y1      = fun(x) 			# clean points
y2      = y1 + nn * std		# noisy points

plt.plot(x, y1, 'b.', x, y2, 'k.')
plt.show()
```

```python
def poly_fit(x, y1, i):
    z = np.polyfit(x, y1, i)
    p = np.poly1d(z)

    xp = np.linspace(-10, 10, 10)
    plt.plot(x, y1, 'b.', xp, p(xp), 'r')
    plt.show()
    return z, p
```

```python
def error(f, x, y1):
    return sum((y1 - f(x) ** 2))
```

```python
errors = list()

for i in range(0, 6):
    z, p = poly_fit(x, y1, i)
    errors.append(error(p, x, y1))
```

```python
plt.plot(list(range(0, len(errors))), errors, color="red",)
plt.show()
```
